```python
# Import the pandas library, which is the primary tool for data analysis
import pandas as pd

# Load the datasets
try:
    df_true = pd.read_csv("data/True.csv")
    df_fake = pd.read_csv("data/Fake.csv")
except FileNotFoundError:
    print("---")
    print("ERROR: Make sure 'True.csv' and 'Fake.csv' are in the 'data/' folder.")
    print("---")

# --- Data Preparation ---

# 1. Create a 'label' column to mark the data
#    0 = Fake News
#    1 = Real News
df_true['label'] = 1
df_fake['label'] = 0

# 2. Combine the two DataFrames into one single DataFrame
#    We use concat() to stack them on top of each other
df = pd.concat([df_true, df_fake], ignore_index=True)

# 3. Shuffle the combined data to mix up the 'true' and 'fake' entries
#    frac=1 means shuffle 100% of the rows.
#    random_state=42 ensures you get the same "random" shuffle every time.
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# --- Initial Analysis ---

# 4. Display the first 5 rows to see what our combined data looks like
print("First 5 rows of the combined dataset:")
print(df.head())

# 5. Get a summary of the data (column names, count, data types)
print("\nDataset Information:")
df.info()

# 6. Check the balance of our labels
print("\nLabel Distribution:")
print(df['label'].value_counts())
```
